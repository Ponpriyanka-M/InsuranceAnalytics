```python
import pandas as pd
import numpy as np
import random
import string
from faker import Faker
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
from imblearn.over_sampling import SMOTE
import nltk

# Download necessary NLTK data
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")

# Initialize Faker for generating synthetic dataset
fake = Faker()
```

```python
def generate_fake_reviews(n=500):
    data = []
    for _ in range(n):
        review_text = fake.sentence(nb_words=random.randint(10, 30))
        rating = random.randint(1, 5)  # 1-5 Star Ratings
        customer_id = fake.uuid4()
        review_id = fake.uuid4()
        service_type = random.choice(["Claim", "Policy Purchase", "Support"])

        # Convert rating to sentiment
        if rating in [1, 2]:
            sentiment_label = "Negative"
        elif rating == 3:
            sentiment_label = "Neutral"
        else:
            sentiment_label = "Positive"

        data.append([review_id, customer_id, review_text, sentiment_label, rating, service_type])

    return pd.DataFrame(data, columns=["Review_ID", "Customer_ID", "Review_Text", "Sentiment_Label", "Rating", "Service_Type"])

# Generate Data
df_feedback = generate_fake_reviews(500)
df_feedback.head()
```

```python
import nltk
import pandas as pd
import numpy as np
import random
import string
from faker import Faker
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
from imblearn.over_sampling import SMOTE


# Download necessary NLTK data
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
# Download the 'punkt_tab' resource
nltk.download('punkt_tab')
```

```python
# Load Stopwords & Initialize Lemmatizer
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = text.translate(str.maketrans("", "", string.punctuation))  # Remove punctuation
    tokens = word_tokenize(text)  # Tokenize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]  # Lemmatize & remove stopwords
    return " ".join(tokens)

# Apply Preprocessing to Review Text
df_feedback["Cleaned_Review"] = df_feedback["Review_Text"].apply(preprocess_text)
df_feedback.head()
```

```python
df_feedback["Sentiment_Label"] = df_feedback["Sentiment_Label"].map({"Negative": 0, "Neutral": 1, "Positive": 2})
```

```python
vectorizer = TfidfVectorizer(max_features=500)  # Extract top 500 features
tfidf_features = vectorizer.fit_transform(df_feedback["Cleaned_Review"]).toarray()

# Convert to DataFrame
tfidf_df = pd.DataFrame(tfidf_features, columns=[f"tfidf_{i}" for i in range(tfidf_features.shape[1])])

# Merge TF-IDF Features
df_feedback = pd.concat([df_feedback, tfidf_df], axis=1)
```

```python
# Tokenize Text for Word2Vec
df_feedback["Tokenized_Review"] = df_feedback["Cleaned_Review"].apply(word_tokenize)

# Train Word2Vec Model
word2vec_model = Word2Vec(sentences=df_feedback["Tokenized_Review"], vector_size=100, window=5, min_count=2, workers=4)

# Function to Convert Words into Vectors
def get_word2vec_features(text):
    tokens = word_tokenize(text)
    vecs = [word2vec_model.wv[word] for word in tokens if word in word2vec_model.wv]
    return np.mean(vecs, axis=0) if vecs else np.zeros(100)

# Apply Word2Vec
df_feedback["Word2Vec_Features"] = df_feedback["Cleaned_Review"].apply(get_word2vec_features)

# Convert Word2Vec Features from List to Columns
word2vec_features = np.vstack(df_feedback["Word2Vec_Features"].values)
word2vec_df = pd.DataFrame(word2vec_features, columns=[f"w2v_{i}" for i in range(word2vec_features.shape[1])])

# Merge Word2Vec Features
df_feedback = pd.concat([df_feedback, word2vec_df], axis=1)
```

```python
df_feedback = pd.get_dummies(df_feedback, columns=["Service_Type"], drop_first=True)
```

```python
# Extract Features and Labels
X = df_feedback.drop(columns=["Review_ID", "Customer_ID", "Review_Text", "Tokenized_Review", "Word2Vec_Features", "Sentiment_Label", "Cleaned_Review"])  # Drop non-numeric columns and "Cleaned_Review"
y = df_feedback["Sentiment_Label"]

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert back to DataFrame
df_balanced = pd.DataFrame(X_resampled, columns=X.columns)
df_balanced["Sentiment_Label"] = y_resampled
```

```python
df_balanced.to_csv("customer_feedback_preprocessed.csv", index=False)
print("✅ Preprocessed dataset saved successfully!")
```

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from imblearn.over_sampling import SMOTE
```

```python
# Load the dataset
df = pd.read_csv("customer_feedback_preprocessed.csv")

# Display first few rows
df.head()
```

```python
# Identify features (X) and labels (y)
X = df.drop(columns=["Sentiment_Label"])  # All columns except Sentiment_Label
y = df["Sentiment_Label"]  # Target labels (0=Negative, 1=Neutral, 2=Positive)
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Split the dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Initialize and train the model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
```

```python
# Print Accuracy Score
print("🚀 Model Accuracy:", accuracy_score(y_test, y_pred))

# Print Classification Report
print(classification_report(y_test, y_pred))
```

```python
# Save the trained model for future use
joblib.dump(model, "sentiment_analysis_model.pkl")
print("✅ Model Saved Successfully!")
```
